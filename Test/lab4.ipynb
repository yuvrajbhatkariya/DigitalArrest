### Streming :-

```python
import re
def stem(word):
    word = word.lower()
    length = len(word)

    if(length > 3):
        pattern = r'^(\w+?)(?:ing|ed|s)$'
        word = re.sub(pattern, r'\1', word)

    return word

def stem_sentence(sentence):
    words = sentence.split()           
    stemmed_words = [stem(word) for word in words]  
    return " ".join(stemmed_words)  


dem = stem("Smiling")
print(dem)


sentences = "Today i am going to sea shores"

print(stem_sentence(sentences))
```

```python
sentence = input("Enter a sentence: ")

result = stem_sentence(sentence)

print("Stemmed sentence:", result)
```
